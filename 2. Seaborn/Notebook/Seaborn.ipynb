```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
print("built-in dataset in sns: ")
pprint(sns.get_dataset_names())
alcol = pd.read_csv("wine.csv", delimiter =";", encoding="utf-8")
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
car = sns.load_dataset("car_crashes")
abalone = pd.read_csv("abalone_1.csv")
traffic = pd.read_csv("traffic.csv", parse_dates=[7])
```

# KDE Plot

Kernel density estimation is the sum of the distribution for each point ad add them to obtain just one KDE plot. The default is Gaussian, but you cal also use other distributions

```python
sns.kdeplot(data= alcol,
            x = "alcohol", 
            color="gold",
            shade=True,
            lw = 4,
            alpha=0.3)

plt.title("KDE univariate plot")
pass  # used  to avoid annoying text
```

```python
sns.kdeplot(data= alcol,
            x = "alcohol",
            y = "sulphates",
            cmap = "coolwarm",
            n_levels= 20) # n of counters

plt.title("KDE 2D")
pass
```

```python

setosa = iris.loc[iris["species"]=="setosa"]
virginica = iris.loc[iris["species"]=="virginica"]

sns.kdeplot(x="sepal_width",y= "sepal_length", data=setosa, cmap="husl", n_levels=20)
sns.kdeplot(x="sepal_width",y= "sepal_length", data=virginica, cmap="inferno", n_levels=20)
plt.legend(["Setosa","virginica"])
plt.title("Multiple KDE")
pass
```

# Histplot-Displot

Is a plot combining all the possible way to visualize the distribution of the distribution onf some variable.

```python
sns.histplot(data =alcol,
            x="alcohol",
            kde=True,
            alpha=1,
            color="firebrick")
plt.title("DisPlot")
pass
```

```python
sns.histplot(data =iris,
            x="sepal_width",
            kde=True,
            hue="species",
            palette="magma",
            lw = 1)
plt.title("Histogram")
pass
```

```python
sns.displot(data=car,
            x="not_distracted",
            kde = True,
            bins = 20,
            color="firebrick",
            edgecolor="black")
plt.title("DisPlot")
pass
```

# Bar Plot

```python
from numpy import mean
sns.barplot(data=abalone,
            y= "Diameter",
            x="Sex" ,
            palette="spring",
            order=["I","F","M"], 
            capsize=0.2, 
            estimator=mean)
plt.title("BarPlot")
pass
```

```python
sns.barplot(data=tips,
            x ="day", 
            y ="total_bill",
            hue="sex",
            palette="autumn")
plt.title("BarPlot")
pass
```

```python
sns.countplot(data=abalone,
              x ="Sex",
              palette= "RdYlGn")
plt.title("BarPlot")
pass
```

# Line Plot

```python
traffic["binary"] = [not x for x in traffic["holiday"] == "None"]
```

```python
sns.lineplot(data=traffic,
            x = "clouds_all",
            y="traffic_volume",
            hue ="binary",
            palette="dark")
plt.title("LinePlot")
pass
```

# Box Plot

```python
sns.boxplot(data=tips,
            x = "day",
            y ="total_bill", 
            hue="sex",
            palette="husl",
            orient = "v")
plt.title("BoxPlot")
pass
```

```python
sns.boxplot(data=tips,x = "sex", y="total_bill", palette="bwr")
sns.swarmplot(data=tips,x = "sex", y="total_bill", palette="magma")
plt.title("BoxPlot")
pass
```

```python
sns.boxplot(data=tips,x="day",y="total_bill", palette="inferno")
sns.stripplot(data=tips,x="day",y="total_bill", palette="inferno")
plt.title("BoxPlot")
pass
```

# Violine Plot

```python
sns.violinplot(data=iris,
               x="species", 
               y="sepal_width",
               palette="Spectral",
               inner="quartile")
plt.title("ViolinePlot")
pass
```

# Scatter Plot 

```python
sns.set_style("white")
sns.scatterplot(x="sepal_length",
                y="petal_length", 
                data=iris,
               palette="inferno", 
                hue="species")
plt.title("ScatterPlot")
pass
```

```python
sns.set_style("dark")
sns.stripplot(data=tips,
              x="day",
              y="total_bill",
             palette="magma",
              jitter=0.0001)
plt.title("StripPlot")
pass
```

```python
sns.set_style("dark")
sns.stripplot(x="day", y="total_bill", data=tips,
              palette="winter_r", 
              dodge=True, 
              hue="sex",
              linewidth= 1,
              marker="d",
              size = 10,
              alpha=0.6)
plt.title("StripPlot")
pass
```

# PairGrid

```python
plot = sns.PairGrid(data=iris, hue="species", palette="autumn", hue_kws={"marker":["d","s","+"]})
plot.map_diag(plt.hist)
plot.map_upper(plt.scatter)
plot.map_lower(sns.kdeplot)
plot.add_legend()
pass
```

```python
l=sns.PairGrid(iris, hue="species")
l.map(plt.scatter)
pass
```

# Facet Grid

```python
tips_facet = sns.FacetGrid(data= tips, 
                           col="time", 
                           row="smoker",
                           aspect=1,
                           palette="Set1")

tips_facet.map(sns.regplot, "total_bill", "tip")
pass
```

```python
tips_facet = sns.FacetGrid(data=tips, 
                           col="time", 
                           row="smoker",
                           aspect=1,
                           palette="Set1")
tips_facet.map(plt.hist, "total_bill", color="firebrick")
pass
```

```python
tips_facet = sns.FacetGrid(tips, 
                           col="time", 
                           aspect=1,
                           hue = "smoker")

tips_facet.map(plt.scatter, "total_bill", "tip")
tips_facet.add_legend()
pass
```
